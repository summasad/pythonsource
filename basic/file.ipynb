```python
# score.txt 읽어온 후 합계와 평균 구한 후 출력
f = open("./score.txt", "r", encoding="utf-8")

scores = []
for num in f: 
    scores.append(int(num))
print(f"평균 : {sum(scores) / len(scores)}")
```

```python
# sample.txt 읽어온 후 합계와 평균 구한 후 
# result.txt 작성
f = open("./sample.txt", "r", encoding="utf-8")
scores = []
for num in f: 
    scores.append(int(num))



f1 = open("./result.txt", "w", encoding="utf-8")
f1.write(f"합계: {sum(scores)}\n")
f1.write(f"평균: {sum(scores)/len(scores)}\n")
f1.close()
```

```python
# info.txt 읽기
# BHI 지수 계산 후 화면에 출력
# [출력결과]
# 타차, 66, 158
# 이름 : 타차
# 몸무게 : 66
# 키 : 158
# BHI : 33.156 (bmi = weight / (height / 100) ** 2)
# 결과 : 과체중

f = open("./info.txt", "r", encoding="utf-8")

for info in f:
    name, weight, height == info.strip().split(", ")
    bmi =  int(weight) / (int(height) / 100) ** 2

    if 25<=bmi:
        result = "과체중"
    elif 18.5 <= bmi:
        result = "정상체중"
    else :
        result ="저체중"
    print(name, weight, height)
    print(f"이름 : {name} ")
    print(f"몸무게 : {weight} ")
    print(f"키 : {height} ")
    print(f"BMI : {bmi} ")
    print(f"결과 : {result} ")
    
```

```python

while True:
    no = int(input("1. 암호화 | 2. 복호화 | 3. 종료"))
    if no == 1:
        # origin 읽은 후 암호화 후 encryption.txt 로 저장
        f = open("./origin.txt", "r", encoding="utf-8")
        content= f.read()

        f1 = open("./encryption.txt", "w", encoding="utf-8")
        for c in content:
            f1.write(chr(ord(c)+100))
        
    elif no == 2:
        # encryption.txt 를 읽은 후 원래 문자대로 화면 출력
        f2 = open("./encryption.txt", "r", encoding="utf-8")

        content= f2.read()
        for i in range(len(content)):
            print(chr(ord(content[i])-100), end="")
    else:
        break
```

#### csv
- csv(comma separated value : 표 형태의 데이터를 쉼표로 구분)

```python
import csv
with open("./sample1.csv") as f:
    reader = csv.reader(f)
    #print(reader)

    next(reader) # 헤더명 제거
    for c in reader:
        print(c)
```

```python
with open("./sample2.csv") as f:
    reader = csv.reader(f, delimiter="|")
    #구분자 | 기준으로 나눠서 가져오기

    next(reader) # 헤더명 제거
    # list 아닌 개별로 가져오기
    for row in reader:
        for col in row:
            print(col)
```

```python
with open("./sample1.csv") as f:
    reader = csv.DictReader(f)
    #print(reader)

    next(reader) # 헤더명 제거
    for c in reader:
        for k, v in c.items():
            print(k, v)
```
