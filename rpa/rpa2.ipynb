```python
import pyautogui as p

# 해상도 확인
size = p.size()
print(size)
print(size[0])
```

```python
# 마우스 조작

# 마우스 좌표 가져오기
position = p.position()
print(position)
print(position.x, position.y)
```

```python
# 마우스 이동 : moveTo(x,y, 움직이는 시간) - 절대좌표

p.moveTo(100,100,0.5)
p.moveTo(200,200,0.5)
p.moveTo(300,300,0.5)
```

```python
# 마우스 이동 : move(x,y, 움직이는 시간), moveRel() -상대좌표
p.moveTo(100,100,0.5)
p.move(300,300,0.25)
print(p.position())
p.move(100,100,0.25)
p.moveRel(200,200,0.25)
```

```python
# 현재 위치에서 클릭

p.click()
#p.click(400,400) : 해당좌표에서 클릭
```

```python
# 마우스 동작 쉬기(시간)
p.sleep(3)

print(p.position())
```

```python
#마우스 클릭(x,y,시간)
p.click(42,12,1)
```

```python
#더블클릭
p.click(clicks=2)
```

```python
#더블클릭
p.doubleClick()
```

```python
#우클릭
p.rightClick()
```

```python
# drag() : 상대위치로 drag

#Point(x=1638, y=519)
p.moveTo(x=1638, y=519)
p.drag(-150,0,0.25)
```

```python
# dragTo() : 절대좌표로 drag
p.moveTo(x=1638, y=519)
p.dragTo(1400,500,0.25)
```

```python
# 양수인 경우 : 위,  음수인 경우 : 아래

#Point(x=1700, y=505) 브라우저 내부
p.moveTo(x=1700, y=505)
p.sleep(2)
p.scroll(-500)
```

```python
#윈도우

window = p.getActiveWindow()
print(window)
print(window.title)
print(window.top, window.left, window.width, window.height)
```

```python
for w in p.getAllWindows():
    print(w)
```

```python
# 특정 창 가져오기
for w in p.getWindowsWithTitle("제목 없음"):
    print(w)
```

```python
# 활성화
w = p.getWindowsWithTitle("제목 없음")[0]
if not w.isActive:
    w.activate()
```

```python
#최대화 되지 않았다면 최대화
if not w.isMaximized:
    w.maximize()
```

```python
#최소화 되지 않았다면 최소화
if not w.isMinimized:
    w.minimize()
```

```python
import pyperclip

w = p.getWindowsWithTitle("제목 없음")[0]
w.activate()
pyperclip.copy("안녕하세요") # 클립보드
p.hotkey("ctrl", "v") #ctrl + V
```

```python
def hangul_write(text):
    pyperclip.copy(text)
    p.hotkey("ctrl","v")
```

```python
w = p.getWindowsWithTitle("제목 없음")[0]
w.activate()
hangul_write("반갑습니다")
hangul_write("hello")
```
